## Carregando o Modelo

```python
model_ner = spacy.load("/content/content/model_ner")
```

```python
rotulos = list(model_ner.get_pipe("ner").labels)
print(rotulos)
```

```python
cores = {
     'B-JURISPRUDENCIA': '#FF595E',
     'B-LEGISLACAO': '#FF924C',
     'B-LOCAL': '#FFCA3A',
     'B-ORGANIZACAO': '#8AC926',
     'B-PESSOA': '#1982C4',
     'B-TEMPO': '#6A4C93',
     'I-JURISPRUDENCIA': '#FF595E',
     'I-LEGISLACAO': '#FF924C',
     'I-LOCAL': '#FFCA3A',
     'I-ORGANIZACAO': '#8AC926',
     'I-PESSOA': '#1982C4',
     'I-TEMPO': '#6A4C93',
     'LOC': '#415a77',
     'MISC': '#415a77',
     'ORG': '#415a77',
     'PER': '#415a77'
}

options = {"ents": rotulos, "colors": cores}
```

## Testando o modelo

```python
with open('/content/Documento Teste.txt', encoding='utf-8') as arch:
  doc = arch.read()
```

```python
#Mostrando o documento
print(doc)
```

```python
#Aplicando modelo
doc_ner = model_ner(doc)
spacy.displacy.render(doc_ner, style="ent", options=options, jupyter=True)
```
